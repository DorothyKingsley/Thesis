```python
from scipy.stats import zscore
import numpy as np
import pandas as pd
df=pd.read_csv("GDA_2016_SA.csv")
#Truncating the numerical values and putting it in a dataframe
df1=df[df.columns[32:]]
#Number of columns 804 after TRUNC(original data) 32 columns
#removing columns which has zeroes
df11=df1.loc[:,(df1 !=0).any(axis=0)]
df11
```

```python
#applying z-score after dropping (all zero) columnms
#converting data frame to a numpy matrix
a=np.array(df11)
#computing zscore
z=zscore(a,axis=1)
#converting to data frame!!!
df2=pd.DataFrame(z)
df2
```

```python
# euclidean distance matrix
from scipy.spatial import distance
D= distance.cdist(df2,df2,'euclidean')
df3=pd.DataFrame(D)
df3.to_csv('GDA_EDistance.csv', index=False)
```

```python
#Similarity matrix function: S[i,j]=1/distance[i,j]
def similarity_matrix(A):
    s = [[0 for x in range(len(A))] for y in range(len(A))] 
    for i in range(len(A)):
        for j in range(len(A)):
            if i==j:
                s[i][j]=0
            elif A[i][j]==0:
                s[i][j]=0
            else:
                s[i][j]=1/(A[i][j])
    return s
#Calculating similarity matrix from euclidean distance
S=similarity_matrix(D)
df4=pd.DataFrame(S)
df4.to_csv('GDA_Similarity.csv', index=False)
```
